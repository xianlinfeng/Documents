```python
# generate the data
f = open("2008",'r')
str = f.read()
f.close()
# read the file

D = list()
for i in range(0,(len(str)+1),3):
    s = str[i:(i+3)]
    if(len(s)==3):
        D.append(int(s))
len(D)
```

```python
from multiprocessing import Pool
# define a Round-Robin data partitioning function
def rr(data,n):
    results = []
    for i in range(n):
        results.append([])
    
    index = 0 
    for i in range(0,len(data),n):
        for j in range(n):
            if(index < len(data)):
                results[j].append(data[index])
                index += 1
    return results

# define a linear search function
def linear_search(data,key):
    """ # this will return the set of positions 
    position = []
    for x in data:
        if x == key:
            position.append(data.index(x))
    return position
    """ 
    count = 0
    for x in data:
        if x == key:
            count += 1
    return count

# define a parallel search function
def paralle_search(data,query,n):
    # data = input data
    # query = a query record
    # n = number of processors
    # p = partioning method
    # s = searching method
    
    print(f"The dataset size is {len(D)}.")
    pool = Pool(processes=n)
    DD = rr(D,n) # partition the data
    
    """ This will return the set of result
    results = []
    for d in DD:
        result = pool.apply_async(linear_search,[d,query])
        output = result.get()
        results.append(output)
    """ 
    
    # this will just return the number of matched result
    count = 0 
    for d in DD:
        result = pool.apply_async(linear_search,[d,query])
        output = result.get()
        count += output
    
    print(f"There are {count} matching results")
    
    return count

# define a parallel search function
def paralle_search(data,query,n):
    # data = input data
    # query = a query record
    # n = number of processors
    # p = partioning method
    # s = searching method
    
    print(f"The dataset size is {len(D)}.")
    pool = Pool(processes=n)
    DD = rr(D,n) # partition the data
    
    """ This will return the set of result
    results = []
    for d in DD:
        result = pool.apply_async(linear_search,[d,query])
        output = result.get()
        results.append(output)
    """ 
    
    # this will just return the number of matched result
    count = 0 
    for d in DD:
        result = pool.apply_async(linear_search,[d,query])
        output = result.get()
        count += output
    
    print(f"There are {count} matching results")
    
    return count
```
