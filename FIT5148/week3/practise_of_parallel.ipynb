```python
from multiprocessing import Pool
import random

# define a Round-Robin data partitioning function
def rr_part(data,n):
    results = []
    for i in range(n):
        results.append([])
    
    index = 0 
    for i in range(0,len(data),n):
        for j in range(n):
            if(index < len(data)):
                results[j].append(data[index])
                index += 1
    return results

# Hash data partitioning
def hash(x,n):
    # define a hash function
    result = x%n
    return result

def hash_part(data,n):
    # define a hash partition function where data is a list type,and n
    # is the number of processers
    dic = {}
    for x in data:
        h = hash(x,n)
        if(h in dic.keys()):
            s = dic[h]
            s.add(x)
            dic[h] = s
        else:
            s = set()
            s.update({x})
            dic[h] = s
    for i in dic.keys():
        lis = list(dic[i])
        dic[i] = lis
    return dic

# range partitioning
def range_part(data,indices):
    # define a range partitioning function which return the subsets of data
    
    result = []
    new_data = list(data)
    new_data.sort()
    indices.append((new_data[-1])+1)
    
    for l in range (len(indices)):
        s = [x for x in new_data if x < indices[l]]
        result.append(s)
        leng = len(s)
        new_data = new_data[leng:]
    return result

# example range_part(D,[40,80])


# Random-unequal data partitioning function

def random_part(data,n):
    
    result = []
    
    """ code start here"""
    dic = {}
    for i in range (n):
        dic[i] = []
        
    
    code = random.randint(0,n)
    for i in data:
        code = random.randint(0,(n-1))
        # dic[code].update(i)
        dic[code].append(i)   
    """ code end here"""
    return dic

#example random_part(D,8)


# define a linear search function
def linear_search(data,key):
    """ # this will return the set of positions 
    position = []
    for x in data:
        if x == key:
            position.append(data.index(x))
    return position
    """ 
    count = 0
    for x in data:
        if x == key:
            count += 1
    return count



# binary search function
def binary_search(data, key):
    position = -1 # not found position
    
    lower = 0
    middle = 0
    upper = len(data)-1
    
    
    """ start code here"""
    while (lower <= upper):
        # calculate middle: the half of lower and upper
        middle = int((lower + upper)/2)
        if key == data[middle]:
            position = middle
            break
        elif middle == upper or middle == lower:
            break
        elif key < data[middle]:
            upper = middle
        elif key > data[middle]:
            lower = middle      
        
    """ end code here"""
    
    return position

# define a parallel search function
def paralle_search_rr_linear(data,query,n):
    # data = input data
    # query = a query record
    # n = number of processors
    # p = partioning method
    # s = searching method
    
    print(f"The dataset size is {len(D)}.")
    pool = Pool(processes=n)
    DD = rr_part(D,n) # partition the data
    
    """ This will return the set of result
    results = []
    for d in DD:
        result = pool.apply_async(linear_search,[d,query])
        output = result.get()
        results.append(output)
    """ 
    
    # this will just return the number of matched result
    count = 0 
    for d in DD:
        result = pool.apply_async(linear_search,[d,query])
        output = result.get()
        count += output
    
    print(f"There are {count} matching results")
    
    return count
```

```python
# another example for hash_partition and binary search
n = 8
query = 9294
pool = Pool(processes=n)

# generate the data
f = open("2006",'r')
str = f.read()
f.close()
# read the file

D = list()
for i in range(0,(len(str)+1),13):
    s = str[i:(i+13)]
    if len(s) != 0:
        D.append(int(s))

data = list(set(D))
DD = hash_part(data,n)

for d in DD.keys():
    ss = DD[d]
    ss.sort()
    DD[d] = ss
```

```python
# connect to the previous section for hash partitioning and binary search
results = []
for d in DD.keys():
    ss = DD[d]
    result = pool.apply_async(binary_search,[ss,query])
    output = result.get()
    results.append(output)    
print(f"The position of the result in each data partition are {results} ")
```
